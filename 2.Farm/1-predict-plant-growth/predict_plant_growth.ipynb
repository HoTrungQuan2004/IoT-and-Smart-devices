```python
import paho.mqtt.client as mqtt
import time
import random
from datetime import datetime

# MQTT Broker settings
broker = "test.mosquitto.org"
port = 1883
topic = "sensor/temperature"

# Set up MQTT publisher
client = mqtt.Client()
client.connect(broker, port, 60)

print("Publishing temperature data...")

for _ in range(10):  # Send 10 temperature values
    temperature = round(random.uniform(20, 35), 2)
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    message = f"{timestamp},{temperature}"  # Ensure correct format

    client.publish(topic, message)
    print(f"Sent: {message}")
    time.sleep(2)  # Wait 2 seconds before sending the next value

print("Publishing complete.")
client.disconnect()
```
